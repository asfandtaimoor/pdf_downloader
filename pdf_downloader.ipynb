```python
import os
download_folder = 'pdf_downloaded'
os.makedirs(download_folder, exist_ok=True)

download_folder_2 = 'pdf_downloaded_count'
os.makedirs(download_folder_2, exist_ok=True)

def get_last_processed_id(folder_path):
    """Returns the highest ID (from filename) in the folder."""
    pdf_files = [f for f in os.listdir(folder_path) if f.endswith('.pdf')]
    if not pdf_files:
        return None

    # Extract numeric part from filenames like 'document_130259.pdf'
    pdf_ids = [int(f.split('_')[1].split('.')[0]) for f in pdf_files]
    
    # Return the maximum ID
    return max(pdf_ids)

if login_response.status_code == 200:
    print("Login successful!")
    data_url = 'https://www.iautocloud.com.cn/maintion/manual.html?orgId=23&brandId=14&carsId=20312&yearId=20313'
    page_response = session.get(data_url)

    if page_response.status_code == 200:
        # Step 4: Parse the HTML
        soup = BeautifulSoup(page_response.content, 'html.parser')

        # Find the script tag that contains pdfData
        script_tags = soup.find_all('script')
        pdf_data = None

        for script in script_tags:
            if script.string and 'var pdfData' in script.string:
                # Use regex to extract the JSON data
                match = re.search(r'var pdfData = (\[.*?\]);', script.string, re.DOTALL)
                if match:
                    pdf_data = match.group(1)  # Get the JSON string
                    break

        # Step 6: Parse the JSON data and extract ids
        # print("pdf_data:", pdf_data)
        if pdf_data:
            pdf_data_dict = json.loads(pdf_data)
            
            # Function to recursively extract IDs
            def extract_ids(data):
                ids = []
                for entry in data:
                    ids.append(entry['id'])  # Add the current ID
                    # If there are children, call the function recursively
                    if 'children' in entry and entry['children']:
                        ids.extend(extract_ids(entry['children']))
                return ids
            
            # Get all IDs
            ids_list = extract_ids(pdf_data_dict)
            print("Count of IDs:", len(ids_list))  
            print("Extracted IDs:", ids_list)
        else:
            print("pdfData not found in the script tags.")
    else:
        print(f"Failed to retrieve the page. Status code: {page_response.status_code}")
else:
    print("Login failed!")
```
